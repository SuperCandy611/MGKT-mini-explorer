```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

DATA_PATH = Path('../data/raw/data.csv')
OUT_PATH  = Path('../reports/age_distribution.png')
DATASET   = 'MGKT Mini Explorer'

# Load
df = pd.read_csv(DATA_PATH)
total_rows = len(df)

# Filter
df_clean = df[(df['age'] >= 13) & (df['age'] <= 80)]
kept_rows = len(df_clean)

# Plot
fig, ax = plt.subplots(figsize=(9, 5))
ax.hist(df_clean['age'], bins=30, color='steelblue', edgecolor='white', linewidth=0.6)
ax.set_xlabel('Age (years)', fontsize=12)
ax.set_ylabel('Count', fontsize=12)
ax.set_title(f'{DATASET} — Age Distribution  (n = {kept_rows:,})', fontsize=13, fontweight='bold')
ax.spines[['top', 'right']].set_visible(False)
plt.tight_layout()

# Save
OUT_PATH.parent.mkdir(parents=True, exist_ok=True)
fig.savefig(OUT_PATH, dpi=150)
plt.show()

# Report
print(f'Total rows loaded : {total_rows:,}')
print(f'Rows kept (13-80) : {kept_rows:,}')
print(f'Figure saved to   : {OUT_PATH.resolve()}')
```
