```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

# 設定支援中文的字型（Windows 環境）
plt.rcParams['font.family'] = ['Microsoft JhengHei', 'Microsoft YaHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# 讀取資料並過濾合理年齡範圍（13–100 歲）
df = pd.read_csv('../data/raw/data.csv')
ages = df['age'].dropna()
ages = ages[(ages >= 13) & (ages <= 100)]

# 繪製年齡分布圖
fig, ax = plt.subplots(figsize=(10, 5))
ax.hist(ages, bins=range(13, 101, 2), color='#4C9BE8', edgecolor='white', linewidth=0.5, zorder=2)

# 樣式設定
ax.set_facecolor('#F7F9FC')
fig.patch.set_facecolor('#F7F9FC')
ax.grid(axis='y', color='white', linewidth=1.2, zorder=1)
ax.set_xlabel('年齡 (Age)', fontsize=13, labelpad=8)
ax.set_ylabel('人數 (Count)', fontsize=13, labelpad=8)
ax.set_title('受測者年齡分布\nAge Distribution of Respondents', fontsize=15, fontweight='bold', pad=14)
ax.xaxis.set_major_locator(mticker.MultipleLocator(5))
ax.spines[['top', 'right']].set_visible(False)

# 標註中位數
median_age = ages.median()
ax.axvline(median_age, color='#E05C5C', linestyle='--', linewidth=1.5, label=f'中位數 Median = {median_age:.0f}')
ax.legend(fontsize=11)

plt.tight_layout()
plt.savefig('../data/age_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
print(f'樣本數 N = {len(ages):,}  |  平均年齡 Mean = {ages.mean():.1f}  |  中位數 Median = {median_age:.0f}')
```
